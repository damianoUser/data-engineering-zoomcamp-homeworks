```python

"""
df = pd.read_csv('https://github.com/DataTalksClub/nyc-tlc-data/releases/download/yellow/yellow_tripdata_2021-01.csv.gz')

Oppure, più visivamente:
prefix = 'https://github.com/DataTalksClub/nyc-tlc-data/releases/download/yellow/'
url = f'{prefix}/yellow_tripdata_2021-01.csv.gz'
df = pd.read_csv(url)
"""

url = 'https://github.com/DataTalksClub/nyc-tlc-data/releases/download/yellow/yellow_tripdata_2021-01.csv.gz'


dtype = {
    "VendorID": "Int64",
    "passenger_count": "Int64",
    "trip_distance": "float64",
    "RatecodeID": "Int64",
    "store_and_fwd_flag": "string",
    "PULocationID": "Int64",
    "DOLocationID": "Int64",
    "payment_type": "Int64",
    "fare_amount": "float64",
    "extra": "float64",
    "mta_tax": "float64",
    "tip_amount": "float64",
    "tolls_amount": "float64",
    "improvement_surcharge": "float64",
    "total_amount": "float64",
    "congestion_surcharge": "float64"
}

parse_dates = [
    "tpep_pickup_datetime",
    "tpep_dropoff_datetime"
]

df = pd.read_csv(
    url,
    # nrows=100,
    dtype=dtype,
    parse_dates=parse_dates
)
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql+psycopg2://root:root@localhost:5432/ny_taxi')

# database://user:password@host:port/database_name

"""
Creo un oggetto che sa:
- quale DBMS usare (Postgres)
- quale driver (psycopg2)
- dove connettersi (host, porta, db)
- con quali credenziali (user, pw).

Stiamo dicendo: questo oggetto è il modo standard di collegarsi a questo DB
"""
```

```python
print(pd.io.sql.get_schema(df, name='yellow_taxi_data', con=engine))

"""
Dato un dataframe (df), produce un comando DDL `CREATE TABLE yellow_taxi_data` (...); usando il dialect SQL specificato nell'engine
Quindi:
df = dataframe
name = nome tabella da creare
con=engine = informazioni di connessione al DB in cui creare la tabella (tra cui, quale dialetto SQL usare)
"""
```

```python
df.head(n = 0).to_sql(name='yellow_taxi_data', con=engine, if_exists='replace')
```

```python
df_iter = pd.read_csv(
    url,
    dtype= dtype,
    parse_dates= parse_dates,
    iterator = True,
    chunksize=100_000,
)
df_iter
```

```python
# df = next(df_iter)
# df[0:1]

# questo codice va iterato manualmente
```

```python
first_chunk = next(df_iter)

first_chunk.head(0).to_sql(
    name="yellow_taxi_data",
    con=engine,
    if_exists="replace"
)

print("Table created")

first_chunk.to_sql(
    name="yellow_taxi_data",
    con=engine,
    if_exists="append"
)

print("Inserted first chunk:", len(first_chunk))

for df_chunk in df_iter:
    df_chunk.to_sql(
        name="yellow_taxi_data",
        con=engine,
        if_exists="append"
    )
    print("Inserted chunk:", len(df_chunk))
```
